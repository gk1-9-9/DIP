# Image Processing with OpenCV

## Prerequisites

Ensure you have the following libraries installed:
- `opencv-python`
- `numpy`
- `matplotlib`

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv.imread("image.png")

# Resize the image
resized_img = cv.resize(img, (1000, 1000))

# Convert to grayscale and HSV
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
hsv_img = cv.cvtColor(img, cv.COLOR_BGR2HSV)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(5, 5))

# Display original image
axs[0, 0].imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
axs[0, 0].set_title('Original Image')
axs[0, 0].axis('off')

# Display resized image
axs[0, 1].imshow(cv.cvtColor(resized_img, cv.COLOR_BGR2RGB))
axs[0, 1].set_title('Resized Image')
axs[0, 1].axis('off')

# Display grayscale image
axs[1, 0].imshow(gray_img, cmap='gray')
axs[1, 0].set_title('Gray Image')
axs[1, 0].axis('off')

# Display HSV image
axs[1, 1].imshow(cv.cvtColor(hsv_img, cv.COLOR_BGR2RGB))
axs[1, 1].set_title('HSV Image')
axs[1, 1].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```
